```python
# 승률/픽률 + champstat
full_IRON = pd.read_csv('forCorrelation/full_IRON_total.csv', index_col=0)
full_BRONZE = pd.read_csv('forCorrelation/full_BRONZE_total.csv', index_col=0)
full_SILVER = pd.read_csv('forCorrelation/full_SILVER_total.csv', index_col=0)
full_GOLD = pd.read_csv('forCorrelation/full_GOLD_total.csv', index_col=0)
full_PLATINUM = pd.read_csv('forCorrelation/full_PLATINUM_total.csv', index_col=0)
full_DIAMOND = pd.read_csv('forCorrelation/full_DIAMOND_total.csv', index_col=0)
full_ABOVE_MASTER= pd.read_csv('forCorrelation/full_ABOVE_MASTER_total.csv', index_col=0)
```

```python
# BIPA 정리
BIPA_IRON = pd.read_csv('forCorrelation/IRON_BIPA.csv', index_col=0)
BIPA_BRONZE = pd.read_csv('forCorrelation/BRONZE_BIPA.csv', index_col=0)
BIPA_SILVER = pd.read_csv('forCorrelation/SILVER_BIPA.csv', index_col=0)
BIPA_GOLD = pd.read_csv('forCorrelation/GOLD_BIPA.csv', index_col=0)
BIPA_PLATINUM = pd.read_csv('forCorrelation/PLATINUM_BIPA.csv', index_col=0)
BIPA_DIAMOND = pd.read_csv('forCorrelation/DIAMOND_BIPA.csv', index_col=0)
BIPA_ABOVE_MASTER= pd.read_csv('forCorrelation/ABOVE_MASTER_BIPA.csv', index_col=0)
```

```python
# json_champ
json_champ = pd.read_csv('forCorrelation/json_champ.csv', index_col=0)
```

```python
# 승률/픽률 + champstat + json_champ
full_IRON = full_IRON.merge(json_champ, how = 'left', on = 'name')
full_BRONZE = full_BRONZE.merge(json_champ, how = 'left', on = 'name')
full_SILVER = full_SILVER.merge(json_champ, how = 'left', on = 'name')
full_GOLD = full_GOLD.merge(json_champ, how = 'left', on = 'name')
full_PLATINUM = full_PLATINUM.merge(json_champ, how = 'left', on = 'name')
full_DIAMOND = full_DIAMOND.merge(json_champ, how = 'left', on = 'name')
full_ABOVE_MASTER = full_ABOVE_MASTER.merge(json_champ, how = 'left', on = 'name')
```

```python
corr_columns = ['winRate', 'pickRate', 'hp', 'hpperlevel', 'mp', 'mpperlevel',
       'movespeed', 'armor', 'armorperlevel', 'spellblock',
       'spellblockperlevel', 'attackrange', 'hpregen', 'hpregenperlevel',
       'mpregen', 'mpregenperlevel', 'crit', 'critperlevel', 'attackdamage',
       'attackdamageperlevel', 'attackspeedperlevel', 'attackspeed', 'hp_18',
       'mp_18', 'armor_18', 'spellblock_18', 'hpregen_18', 'mpregen_18',
       'crit_18', 'attackdamage_18', 'attackspeed_18', 'style', 'difficulty',
       'damageType', 'damage', 'durability', 'crowdControl', 'mobility',
       'utility', 'roles']
```

```python
def corr(df, position, rate):
    df = df[df['position'] == position].sort_values(by=[rate], ascending=True)
    df_corr_data = df[corr_columns]
    df_corr = df_corr_data.corr()
    df_corr = df_corr[rate].sort_values(ascending =False)
    df_corr = pd.DataFrame(df_corr)
    return df_corr
```

```python
IT_corr = corr(full_IRON, 'T', 'pickRate')
IT_corr.rename(columns = {'pickRate':'IT'},inplace=True)
IM_corr = corr(full_IRON, 'M', 'pickRate')
IM_corr.rename(columns = {'pickRate':'IM'},inplace=True)
IA_corr = corr(full_IRON, 'A', 'pickRate')
IA_corr.rename(columns = {'pickRate':'IA'},inplace=True)
IS_corr = corr(full_IRON, 'S', 'pickRate')
IS_corr.rename(columns = {'pickRate':'IS'},inplace=True)
IJ_corr = corr(full_IRON, 'J', 'pickRate')
IJ_corr.rename(columns = {'pickRate':'IJ'},inplace=True)
```
